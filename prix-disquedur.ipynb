# Graphique du cout d'un disque dur

Cout du disque dur en fonction de la capacité de stockage.

## Modèle

Disque dur Seagate BarraCuda Compute

## Source

[Canada Computers](https://www.canadacomputers.com/product_info.php?cPath=15_1086_210&item_id=130405) Le 23 février 2021. **Aucun rabais
n'a été pris en compte.**

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import linregress

plt.style.use("seaborn")
```

## Données

```python
# Les clés sont des capacités en TB
# Les valeurs sont des prix en CAD
prix = {
    "1": 59.99,
    "2": 74.99,
    "4": 109.99,
    "6": 199.99,
    "8": 209.99,
}
```

```python
x, y = [], []
for k, v in prix.items():
    x.append(k)
    y.append(v)
```

```python
x, y
```

```python
for i in range(len(x)):
    x[i] = int(x[i])
```

```python
x, y = np.array(x), np.array(y)
```

## Pente

```python
slope, intercept, r, p, *other = linregress(x,y)
```

```python
linregress(x,y)
```

## Pente entre 1TB et 4TB

```python
slope2, intercept2, r2, p2, *other = linregress(x[0:3],y[0:3])
```

## Graphique

```python
plt.scatter(x,y)
plt.plot(x, x*slope + intercept, label="Toutes les capacités")
plt.plot(x, x*slope2 + intercept2, label="Capacités de 1TB à 4TB")
plt.ylabel("Prix en CAD")
plt.xlabel("Capacité en TB")
plt.legend()
plt.savefig("prix-capacite.pdf")
```
